<h3>Data Reading and Elementary Cleaning

```python
import pandas as pd
import numpy as np
```

```python
#read global csv
data_global = pd.read_excel('globalterrorismdb_0718dist.xlsx')
```

```python
#read 1993 data
data_1993 = pd.read_excel('gtd1993_0718dist.xlsx')
```

```python
data_global.set_index('eventid',inplace=True)
data_1993.set_index('eventid',inplace=True)
```

```python
type(data_1993.eventid[2])
```

```python
#join the two dataframe
data_global.join(data_1993,on='eventid',how='outer',sort=True)
```

```python
data = data_global.append(data_1993)
data
```
